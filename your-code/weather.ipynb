## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
data=pd.read_csv("weather-raw.csv")
```

```python
data
```

```python
#Aplicamos melr para poder limpiar bien la data
data2=data.melt(id_vars=['id', 'year', 'month', 'element'] ,var_name="days" , value_name= "weather")
```

```python
#vemos que hay muchos valores nulos. 
data.isnull().sum()
```

```python
#limpiamos filas eliminando valores nulos en el dataset original
data2.dropna(inplace=True)
```

```python
#comprobamos
data2.head(3)
```

```python
#limpiamos columnas, dejando solo las que nos interesan con "usefull data"
data2.drop(columns=["id"], inplace=True)
```

```python
#comprobamos
data2.head(3)
```

```python
#Reordenamos columnas para que los valores se muestres de mes1 a mes31.
data2.sort_values("month", ascending=True, inplace=True)
```

```python
#comprobamos
data2.head(3)
```

```python
#Reseteamos el index (dudo de si se haría ahora o más adelante, ero creo que para trabajar valores me servirá tenerlo ordenado en este punto)
data2.reset_index(inplace=True)
```

```python
#comprobamos
data2.head(3)
```

```python
#eliminamos columna generada del índice antiguo
data2.drop(columns=["index"], inplace=True)
```
